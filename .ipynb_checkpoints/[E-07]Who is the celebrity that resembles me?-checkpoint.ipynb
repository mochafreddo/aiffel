ÏûêÎ•∏ Ïù¥ÎØ∏ÏßÄÎäî PILÏùò `Image.fromarray`Î•º ÌÜµÌï¥ÏÑú PIL ImageÎ°ú Î≥ÄÌôòÌïú Îí§Ïóê Ï†ÄÏû•ÏùÑ Ìï¥ÎëêÎ©¥ Îã§ÏñëÌïú Ïû¨ÎØ∏ÏûàÎäî ÏãúÍ∞ÅÌôîÎ•º ÏãúÎèÑÌï¥ Î≥¥Îäî Îç∞ ÎèÑÏõÄÏù¥ Îê† Í≤ÉÏûÖÎãàÎã§.

````python
import matplotlib.pyplot as plt
import face_recognition
import os
````

````python
def get_gropped_face(image_file):
    """
    A function that cuts and sprays only the face area when passing the image file path as a parameter.

    :param image_file: Path of the image
    :return: Face image contents as numpy array
    """
    image = face_recognition.load_image_file(image_file)
    face_locations = face_recognition.face_locations(image)
    a, b, c, d = face_locations[0]
    cropped_face = image[a:c, d:b, :]

    return cropped_face
````

### Step3. ÏñºÍµ¥ ÏòÅÏó≠Ïùò ÏûÑÎ≤†Îî© Ï∂îÏ∂úÌïòÍ∏∞

---

Ïù¥Ï†ú ÏñºÍµ¥ ÏòÅÏó≠ÏúºÎ°ú ÏûòÎ¶∞ Ïù¥ÎØ∏ÏßÄÎì§Ïù¥ Ïó¨Îüø ÏÉùÍ≤ºÏùÑ ÌÖêÎç∞Ïöî. Í∑∏ Ïù¥ÎØ∏ÏßÄÎ•º `face_recognition` ÏùÑ ÏÇ¨Ïö©Ìï¥ÏÑú Ïù¥ÎØ∏ÏßÄÎ•º Î∂àÎü¨Ïò§Í≥† ÏûÑÎ≤†Îî©ÏùÑ Ï∂îÏ∂úÌï¥ Î¥ÖÏãúÎã§. Îã§Î•∏ ÏÇ¨ÎûåÏùò ÏûÑÎ≤†Îî©ÏùÄ Íµ¨Î∂ÑÌï† Ïàò ÏûàÍ≤å Îî∞Î°ú Ï†ÄÏû•Ìï¥ Ï£ºÏÑ∏Ïöî. Ïù¥Ï†ÑÏóê ÎßåÎì§ÏóàÎçò Îã§Ïùå Ìï®ÏàòÎì§ÏùÑ ÌôúÏö©ÌïòÎ©¥ ÏÜêÏâΩÍ≤å ÏàòÌñâ Í∞ÄÎä•Ìï©ÎãàÎã§.

- `def get_face_embedding(face)` : ÏñºÍµ¥ÏòÅÏó≠ÏúºÎ°úÎ∂ÄÌÑ∞ ÏñºÍµ¥ ÏûÑÎ≤†Îî© Î≤°ÌÑ∞Î•º Íµ¨ÌïòÎäî Ìï®Ïàò
- `def get_face_embedding_dict(dir_path)` : ÎîîÎ†âÌÜ†Î¶¨ ÏïàÏóê ÏûàÎäî Î™®Îì† Ïù¥ÎØ∏ÏßÄÏùò ÏûÑÎ≤†Îî© ÎîïÏÖîÎÑàÎ¶¨Î•º Íµ¨ÌïòÎäî Ìï®Ïàò

````python
def get_face_embedding(face):
    """
    Function to find face embedding vector from face area.

    :param face: Face image contents as numpy array
    :return: A list of 128-dimensional face encodings (one for each face in the image)
    """
    return face_recognition.face_encodings(face)
````

````python
def get_face_embedding_dict(dir_path):
    """
    A function to get the embedding dictionary of all images in a directory.

    :param dir_path: path containing the image
    :return: embedding_dict{image_file_name: (128-dimensional face encodings)}
    """
    file_list = os.listdir(dir_path)
    embedding_dict = {}

    for file in file_list:
        img_path = os.path.join(dir_path, file)

        # Î¨∏Ï†úÍ∞Ä ÏûàÎäî Ïù¥ÎØ∏ÏßÄÎ•º Í±∏Îü¨ÎÇ¥Í∏∞ ÏúÑÌï¥ ÏòàÏô∏Ï≤òÎ¶¨Î•º ÎëêÍ≥† Î¨∏Ï†úÍ∞Ä ÏûàÎäî Ïù¥ÎØ∏ÏßÄ ÌååÏùºÎ™ÖÏùÑ Ï∂úÎ†•
        try:
            face = get_gropped_face(img_path)
        except:
            print(img_path)

        embedding = get_face_embedding(face)
        if len(embedding) > 0:  # ÏñºÍµ¥ÏòÅÏó≠ faceÍ∞Ä Ï†úÎåÄÎ°ú detectÎêòÏßÄ ÏïäÏúºÎ©¥ len(embedding)==0 Ïù∏ Í≤ΩÏö∞Í∞Ä Î∞úÏÉùÌïòÎØÄÎ°ú
            # os.path.splitext(file)[0]ÏóêÎäî Ïù¥ÎØ∏ÏßÄÌååÏùºÎ™ÖÏóêÏÑú ÌôïÏû•ÏûêÎ•º Ï†úÍ±∞Ìïú Ïù¥Î¶ÑÏù¥ Îã¥ÍπÅÎãàÎã§.
            embedding_dict[os.path.splitext(file)[0]] = embedding[0]

    return embedding_dict
````

### Step4. Î™®ÏùÄ Ïó∞ÏòàÏù∏Îì§Í≥º ÎπÑÍµêÌïòÍ∏∞

---

Î™®ÏùÄ Ïó∞ÏòàÏù∏ ÏÇ¨ÏßÑÎì§Í≥º ÏûêÏã†Ïùò ÏñºÍµ¥ ÏÇ¨ÏßÑÏóêÏÑú Ï∂îÏ∂úÌïú ÏûÑÎ≤†Îî© Î≤°ÌÑ∞Îì§ Í∞ÑÏùò Í±∞Î¶¨Î•º ÎπÑÍµêÌï¥ Î≥¥Í≥† Í∞ÄÏû• ÏûëÏùÄ Ïó∞ÏòàÏù∏Ïù¥ ÎàÑÍµ¨Ïù∏ÏßÄ Ï∞æÏïÑÎ¥ÖÏãúÎã§!

Ïù¥Ï†Ñ Ïä§ÌÖùÏóêÏÑú ÎßåÎì§ÏóàÎçò `get_nearest_face(name, top=5)` Ìï®ÏàòÎì§ÏùÑ ÏùëÏö©ÌïòÎ©¥ ÏÜêÏâΩÍ≤å ÏàòÌñâ Í∞ÄÎä•Ìï©ÎãàÎã§.

````python
import numpy as np
````

````python
def get_distance(name1, name2):
    """
    Calculate the distance (L2 distance) between two people's face embeddings.

    :param name1: First person
    :param name2: Second person
    :return: float or ndarray (Norm of the matrix or vector(s).)
    """
    return np.linalg.norm(embedding_dict[name1]-embedding_dict[name2], ord=2)
````

````python
# name1Í≥º name2Ïùò Í±∞Î¶¨Î•º ÎπÑÍµêÌïòÎäî Ìï®ÏàòÎ•º ÏÉùÏÑ±ÌïòÎêò, name1ÏùÄ ÎØ∏Î¶¨ ÏßÄÏ†ïÌïòÍ≥†, name2Îäî Ìò∏Ï∂úÏãúÏóê Ïù∏ÏûêÎ°ú Î∞õÎèÑÎ°ù Ìï©ÎãàÎã§.
def get_sort_key_func(name1):
    def get_distance_from_name1(name2):
        return get_distance(name1, name2)
    return get_distance_from_name1
````

````python
def get_nearest_face(name, top=5):
    """
    Find an image that most resembles a specific image.

    :param name: The filename of the image of the person to look for.
    :param top: Number of resemblances to find.
    """
    sort_key_func = get_sort_key_func(name)
    sorted_faces = sorted(embedding_dict.items(),
                          key=lambda x: sort_key_func(x[0]))

    for i in range(top+1):
        if i == 0:  # Ï≤´Î≤àÏß∏Î°ú ÎÇòÏò§Îäî Ïù¥Î¶ÑÏùÄ ÏûêÍ∏∞ ÏûêÏã†Ïùº Í≤ÉÏù¥ÎØÄÎ°ú Ï†úÏô∏Ìï©ÏãúÎã§.
            continue
        if sorted_faces[i]:
            print('ÏàúÏúÑ {} : Ïù¥Î¶Ñ({}), Í±∞Î¶¨({})'.format(
                i, sorted_faces[i][0], sort_key_func(sorted_faces[i][0])))
````

### Step5. Îã§ÏñëÌïú Ïû¨ÎØ∏ÏûàÎäî ÏãúÍ∞ÅÌôî ÏãúÎèÑÌï¥ Î≥¥Í∏∞

---

ÎπÑÎã® ÎãÆÏùÄ Íº¥ Ïó∞ÏòàÏù∏ÏùÑ Ï∞æÍ∏∞Îßå Ìï† Í≤ÉÏù¥ ÏïÑÎãàÎùº, Í∑∏Îì§Ïùò ÏñºÍµ¥ ÏÇ¨ÏßÑÏùÑ ÌôúÏö©ÌïòÏó¨ Ïû¨ÎØ∏ÏûàÎäî ÏãúÍ∞ÅÌôîÎ•º ÏãúÎèÑÌï¥ Î≥¥ÏãúÍ∏∞ Î∞îÎûçÎãàÎã§.  
tkinterÎÇò flaskÎ•º ÌôúÏö©Ìï¥ÏÑú ÎãÆÏùÄÍº¥ÏùÑ Ï∞æÏùÑ Ïù¥ÎØ∏ÏßÄÎ•º ÏóÖÎ°úÎìúÌïú ÌõÑ Í∑∏ Ïù¥ÎØ∏ÏßÄÏôÄ ÎãÆÏùÄ ÏñºÍµ¥ÏùÑ Ï∞æÎäî ÌòïÌÉúÎ°ú Íµ¨ÌòÑÌï¥ Î≥¥Îäî Í≤ÉÎèÑ Ïû¨ÎØ∏ÏûàÏùÑ Í≤ÉÏûÖÎãàÎã§.  
Ïó¨Îü¨Î∂ÑÎì§Ïùò Ïû¨ÏπòÍ∞Ä ÎÑòÏπòÎäî ÏûëÌíàÎì§ÏùÑ Í∏∞ÎåÄÌï©ÎãàÎã§. üòé

````python
from sklearn import manifold
from mpl_toolkits.mplot3d import Axes3D  # 3Ï∞®Ïõê Í∑∏ÎûòÌîÑÎ•º ÏúÑÌï¥
````

````python
# t-SNEÏùò Îç∞Ïù¥ÌÑ∞Î°ú ÏÇ¨Ïö©ÌïòÍ∏∞ ÏúÑÌï¥ numpyÏùò ndarrayÎ°ú Î≥ÄÌôò
X = np.array(list(embedding_dict.values()))
````

````python
# t-SNE embedding
tsne = manifold.TSNE(n_components=3,  # 3Ï∞®Ïõê ÏãúÍ∞ÅÌôîÎ•º ÏúÑÌï¥ 3ÏúºÎ°ú ÏÑ§Ï†ï
                     init='pca',
                     random_state=0)
X_tsne = tsne.fit_transform(X)
````

````python
# 3Ï∞®Ïõê ÏÇ∞Ï†êÎèÑ (3D Scatter plot) Í∑∏Î¶¨Í∏∞
plt.rcParams["figure.figsize"] = (20, 20)  # figureÏùò ÏÇ¨Ïù¥Ï¶à
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')  # 3D axesÎ•º ÎßåÎì§Í∏∞ ÏúÑÌï¥
for i in range(X_tsne.shape[0]):
    ax.scatter(X_tsne[i, 0], X_tsne[i, 1], X_tsne[i, 2], marker='o', s=150)
````
