### Step 2. 얼굴 검출 & 랜드마크 검출 하기  
---  
오늘 배운 dlib을 이용해서 얼굴의 bounding box 위치와 landmark의 위치를 찾아주세요.  

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import os
import dlib  # face detector
```

```python
# load the image
my_image_path = os.getenv('HOME')+'/camera_sticker/images/image.jpg'
img_bgr = cv2.imread(my_image_path)

img_bgr = cv2.resize(img_bgr, (595, 599))  # resize
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)  # convert to rgb

# store images for printing separately
img_show = img_rgb.copy()

# 1. img_bgr : original resized image
# 2. img_rgb : converted image
# 3. img_show : image for output
```

```python
# image check
plt.imshow(img_rgb)
plt.show()
```

```python
# declare detector
detector_hog = dlib.get_frontal_face_detector()
```

```python
# using a detector to extract the bounding box of the face
# dlib takes an rgb image as input
dlib_rects = detector_hog(img_rgb, 1)  # (image, num of img pyramid)
```

```python
# draw a green square on the found face areas
for dlib_rect in dlib_rects:
    l = dlib_rect.left()
    t = dlib_rect.top()
    r = dlib_rect.right()
    b = dlib_rect.bottom()

    # draw a green square
    cv2.rectangle(img_rgb, (l, t), (r, b),
                  (0, 255, 0), 2, lineType=cv2.LINE_AA)
```

```python
# image check
plt.imshow(img_rgb)
plt.show()
```

```python
# load the landmark model
model_path = os.getenv(
    'HOME')+'/camera_sticker/models/shape_predictor_68_face_landmarks.dat'
landmark_predictor = dlib.shape_predictor(model_path)
```

```python
# predict landmarks
list_landmarks = []
for dlib_rect in dlib_rects:  # face areas
    points = landmark_predictor(img_show, dlib_rect)
    list_points = list(map(lambda p: (p.x, p.y), points.parts()))
    list_landmarks.append(list_points)
```

```python
# print landmarks to the image
for landmark in list_landmarks:
    for idx, point in enumerate(landmark):
        cv2.circle(img_rgb, point, 2, (0, 255, 255), -1)  # yellow
```

```python
# image check
plt.imshow(img_rgb)
plt.show()
```

### Step 3. 스티커 적용 위치 확인하기  
---  
고양이 수염이 적용 될 볼 위치를 landmark를 사용해서 계산해 주세요.  
![](https://aiffelstaticprd.blob.core.windows.net/media/original_images/E-8-8.png)  
볼은 '턱', '눈', '입', '코' 사이에 있다는 점을 이용해 주세요. 비율로 계산하는 방법 잊지 않으셨죠?  
  
_[hint] 2, 36, 48, 30 번 인덱스를 이용_  

```python
# (x, y) and (w, h) setting
# 힌트와 달리 코의 좌표를 중심으로 위치를 잡았다.
# 수염 크기는 크게 하고 싶어서 스티커의 크기는 얼굴 영역과 같게 잡는다.

for dlib_rect, landmark in zip(dlib_rects, list_landmarks):
    x = landmark[30][0]-dlib_rect.width()//2
    y = landmark[30][1]-dlib_rect.height()//2
    w = dlib_rect.width()
    h = dlib_rect.height()
```

```python
# check (x, y)
print(f'(x, y) = ({x}, {y})')
```

```python
# read the sticker image
sticker_path = os.getenv('HOME')+'/camera_sticker/images/cat-whiskers.png'
img_sticker = cv2.imread(sticker_path)
img_sticker = cv2.resize(img_sticker, (w, h))
```

```python
# adjust (x, y) -> skip
# 코의 좌표를 중심으로 이미지를 위치하기 때문에 좌표를 조정할 일이 없다.
# refined_x = x
# refined_y = y
```

```python
# 고양이 수염 이미지는 얼굴 영역의 안에 들어가기 때문에 이미지 바깥으로 벗어날 일이 없다.
# 그러므로 좌표가 음수가 되는 예외처리를 생략한다.

# 라고 생각했는데 얼굴 각도에 따라 코의 위치가 달라 바깥으로 벗어날 수가 있더라.
# 게다가 얼굴 영역 자체가 이미지 밖으로 삐져나가는 경우가 생겼다.
```

### Step 4. 스티커 적용하기  
---  
(1) 오늘 배운 **`np.where`** 를 사용해서 스티커를 적용해 주세요.  
  
(2) 스티커를 조금 더 자연스럽게 보이게 하려면 어떻게 해야 할까요? 스티커 뒤로 원본 이미지가 같이 보이도록 만들어 봅시다.  
  
_[hint] opencv 의 **`cv2.addWeighted()`** 를 참고하세요._  
![](https://aiffelstaticprd.blob.core.windows.net/media/images/E-8-21.max-800x600_10ru2yB.png)  

```python
# apply the stickers to the image
sticker_area = img_rgb[y:y+img_sticker.shape[0], x:x+img_sticker.shape[1]]
img_rgb[y:y+img_sticker.shape[0], x:x+img_sticker.shape[1]
        ] = np.where(img_sticker == 255, sticker_area, img_sticker).astype(np.uint8)
# 고양이수염 이미지는 투명한 부분의 값이 255 이므로 실습에서 0 으로 주었던 값을 255로 수정
```

```python
# check image
plt.imshow(img_rgb)
plt.show()
```

```python
# addWeighted()를 위한 복사본
img_show_copy = img_show.copy()
```

```python
# apply the stickers to the img_show
sticker_area = img_show[y:y+img_sticker.shape[0], x:x+img_sticker.shape[1]]
img_show[y:y+img_sticker.shape[0], x:x+img_sticker.shape[1]
         ] = np.where(img_sticker == 255, sticker_area, img_sticker).astype(np.uint8)
```

```python
# addWeighted() 적용
# 원본과 스티커를 합성한 이미지를 반씩 섞어 스티커만 흐려지게 했다.
img_show = cv2.addWeighted(img_show, 0.5, img_show_copy, 0.5, 0)
```

```python
# print the final result
plt.imshow(img_show)
plt.show()
```

#### (1) 얼굴 각도에 따라 스티커가 어떻게 변해야할까요?
> 얼굴 각도에 따라 스티커 역시 변해야 할 것 같다.  
> 얼굴을 옆으로 돌렸을 때 거기에 맞게 좌우 이미지의 비율이 변해야 자연스럽게 적용될 수 있기 때문이다.

#### (2) 멀리서 촬영하면 왜 안될까요? 옆으로 누워서 촬영하면 왜 안될까요?
> 멀리서 촬영하면 얼굴이 명확하게 나오지 않아 데이터가 부족해 제대로 찾지 못하는 것 아닐까 생각해본다.  
> 옆으로 누워서 촬영한 것과 대각선으로 기울인 얼굴도 찾지 못하는 것 같다. 물론 한 장씩만 실험해봤지만.  

#### (3) 실행 속도가 중요할까요?
> 중요하다. 촬영하기 전에도 피사체가 정지한 상태가 아니기 때문에 실시간으로 타겟에 맞출 수 있어야 하므로. 그래야 원하는 타이밍에 스티커가 위치한 사진을 촬영할 수 있다.

#### (4) 스티커앱을 만들 때 정확도가 얼마나 중요할까요?
> 스티커앱을 사용하는 사용자들이 적절한 위치에 스티커를 놓고 이미지를 촬영하기 원하기 때문에 정확도는 중요하다. 만일 엉뚱한 곳에 스티커가 위치하는 일이 많아지거나 아예 얼굴을 탐색하지 못 한다면 이용자가 없을 것이다.
